```python
coviddf2 = coviddf2.drop(['deaths_unvax','deaths_pvax','deaths_fvax', 'deaths_boost'], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
```

```python
#train test split
X = coviddf2.drop('deaths_new',axis=1)
y = coviddf2['deaths_new']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, shuffle = False )
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
linreg = LinearRegression()
decreg = DecisionTreeRegressor()
```

```python
#Linear Regression model
linreg.fit(X_train, y_train)
preds_lin = linreg.predict(X_test)
```

```python
decreg.fit(X_train,y_train)
preds_dec = decreg.predict(X_test)
```

```python
#overfitted model
preds_train_dec = decreg.predict(X_train)
r2_score(y_train,preds_train_dec)
```

```python
r2_score(y_test,preds_lin)
```

```python
r2_score(y_test,preds_dec)
```
